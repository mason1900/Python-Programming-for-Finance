**Exercise 0**

*Univariate Regression*

Load the data in *univReg.csv*. Use regression, with specially chosen basis functions, to fit a curve. Hint: sine and exponential.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm

%matplotlib inline

df = pd.read_csv('./univReg.csv')
```

```python
df.head()
```

```python
x = df['indep']
y = df['target']

plt.plot(x, y, 'b')
plt.grid(True)
plt.xlabel('indep')
plt.ylabel('target')
plt.show()
```

```python
matrix = np.zeros((8 + 1, len(x)))    
matrix[8, :] = np.sin(x)*np.exp(x)
matrix[7, :] = np.sin(x)
matrix[6, :] = np.exp(x)
matrix[5, :] = x**5
matrix[4, :] = x**4
matrix[3, :] = x**3
matrix[2, :] = x**2
matrix[1, :] = x
matrix[0, :] = 1

reg = np.linalg.lstsq(matrix.T, y,rcond=None)[0]
print("regression parameters:",reg)        # optimal parameters for the single basis functions:
print("regression shape:", reg.shape)
print ("matrix shape:", matrix.shape)
ry = np.dot(reg, matrix)
plt.plot(x, y, 'b', label='target')
plt.plot(x, ry, 'r.', label='regression')
plt.legend(loc=0)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title("Regression via least-squares function")
plt.show()
```

```python
np.allclose(ry, y), (np.sum(y - ry) ** 2) / len(x)
```

*Multivariate Regression*

Load the data in *mulitReg.csv* and treat it as the `Z` value as in our multivariate regression example. Find appropriate basis functions (hint: plot it...also, sine plus a power of $y$) using a multivariate regression on the meshgrid defined below.

```python
import numpy as np
x = np.linspace(0, 10, 20)
y = np.linspace(0, 10, 20)
X, Y = np.meshgrid(x, y)
```

```python
df2 = pd.read_csv('./multiReg.csv')
df2.shape
df2.head()
```

```python
Z = df2
```

```python
import matplotlib

# For unknown reasons, the IPython backend for Matplotlib does not work well on my computer with default settings
# This is a fix.
%config InlineBackend.print_figure_kwargs = {'bbox_inches':None}
%config InlineBackend.rc = {'font.size': 10, 'figure.figsize': (16.0, 9.0), 'figure.facecolor': (1, 1, 1, 0), 'figure.subplot.bottom': 0.125, 'figure.edgecolor': (1, 1, 1, 0), 'figure.dpi': 72}
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl

fig = plt.figure(figsize=(16, 9))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, rstride=2, cstride=2, cmap=mpl.cm.coolwarm, 
       linewidth=0.5, antialiased=True)
#  X, Y, Z - data values as 2D arrays
#  stride, cstride - array row and column stride (step size)
#  cmap - a color map for surface pathes
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
# title: Function with two parameters
```

```python
x2 = X.flatten()
y2 = Y.flatten()
z2 = Z.values.flatten()
matrix = np.zeros((len(x2), 5 + 1))
matrix[:, 5] = np.sin(x2)
matrix[:, 4] = y2 ** 4
matrix[:, 3] = y2 ** 3
matrix[:, 2] = y2 ** 2
matrix[:, 1] = y2
matrix[:, 0] = 1
```

```python
import statsmodels.api as sm
model = sm.OLS(z2, matrix).fit()
model.summary()
```

According to the results above, the appropriate basis function is likely to be $sin(x)+0.05*y^3$

```python
a = model.params
rz2 = np.dot(a, matrix.T)
np.allclose(rz2, z2)
```

```python
def reg_func(a, x, y):
    f5 = a[5] * np.sin(x)
    f4 = a[4] * y ** 4
    f3 = a[3] * y ** 3
    f2 = a[2] * y ** 2
    f1 = a[1] * y
    f0 = a[0] * 1
    return (f5 + f4 + f3 + f2 + f1 + f0)
RZ = reg_func(a, X, Y)
fig = plt.figure(figsize=(9, 6))
ax = fig.gca(projection='3d')
surf1 = ax.plot_surface(X, Y, Z, rstride=2, cstride=2, cmap=mpl.cm.coolwarm,
        linewidth=0.5, antialiased=True)
surf2 = ax.plot_wireframe(X, Y, RZ, rstride=2, cstride=2, label='regression')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.legend()
fig.colorbar(surf, shrink=0.5, aspect=5)
```
