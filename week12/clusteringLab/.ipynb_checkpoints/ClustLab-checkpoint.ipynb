*Logistic Regression*

1. Load the data from "linsep-train.csv" and examine its contents and note the label column. Use it to train a logistic regression classifier.

2. Load the data from "linsep-test.csv" and examine its contents. Apply your classifier to predict class labels, then compare to the true class labels in the "linsep-test.csv" file. What was your *accuracy*? Your *precision*? Your *recall*?

```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import precision_recall_curve
from bokeh.plotting import figure, show, output_notebook
from sklearn.metrics import (precision_score, recall_score, accuracy_score)


output_notebook()

df_train = pd.read_csv('linsep-train.csv')
print(df_train.shape)
df_train.head()
```

```python
logreg = linear_model.LogisticRegression()
X = df_train['x'].values 
Y = df_train['y'].values 
features_train = np.array([X,Y]).transpose()
T_train = df_train['class'].values
logreg.fit(features_train,T_train)
```

```python
df_test = pd.read_csv('linsep-test.csv')
print(df_test.shape)
df_test.head()
```

```python
X_test = df_test['x'].values 
Y_test = df_test['y'].values 
features_test = np.array([X_test,Y_test]).transpose()
pred = logreg.predict_proba(features_test)
T_test = df_test['class'].values

# ROC Curve
precision, recall, thresholds = precision_recall_curve(T_test,pred[:,1])
p1 = figure(plot_width=400, plot_height=400)
p1.line(recall,precision)
show(p1)
```

```python
# Predicted labels
T_pred = logreg.predict(features_test)
print('Accuracy score :', accuracy_score(T_test, T_pred))
print('Precision score :', precision_score(T_test, T_pred))
print('Recall score :', recall_score(T_test, T_pred))
```

2. Add new columns to your `DataFrame` so that the resulting dataset is *linearly separable*, and then separate the resulting dataset using logistic regression or an SVC.

```python
df = pd.read_csv('landlocked.csv')
df.describe()
```

```python
import matplotlib.pyplot as plt


%matplotlib inline
# For unknown reasons, the IPython backend for Matplotlib does not work well on my computer with default settings
# This is a fix.
%config InlineBackend.print_figure_kwargs = {'bbox_inches':None}
%config InlineBackend.rc = {'font.size': 10, 'figure.figsize': (10.0, 10.0), 'figure.facecolor': (1, 1, 1, 0), 'figure.subplot.bottom': 0.125, 'figure.edgecolor': (1, 1, 1, 0), 'figure.dpi': 72}

x = df['x'].values
y = df['y'].values
plt.figure(figsize=(7, 7))
plt.plot(x, y, 'bo')
plt.show()
```

```python
z = [(x[i]**2 + y[i]**2) for i in range(len(x))] 
mark = [1 if z[i]<2 else 0 for i in range(len(z))]

scatter_x = np.array(x)
scatter_y = np.array(y)
group = np.array(mark)
cdict = {0: 'blue', 1: 'red'}

plt.figure(figsize=(10, 10))
for g in np.unique(group):
    ix = np.where(group == g)
    plt.scatter(scatter_x[ix], scatter_y[ix], c = cdict[g], label = g, s = 100)
plt.legend()
plt.show()
```

The correct label is not given in the original dataset. Here I manually give them labels in variable `mark`. According to the graph above, the label I gave seems to be correct.

The following is the clustering algorithm DBSCAN.

```python
# Referenced code from sklearn document.
from sklearn.cluster import DBSCAN
from sklearn import metrics

combined = np.vstack((scatter_x, scatter_y)).T
X = combined
labels_true = np.array(mark)
db = DBSCAN(eps=0.4, min_samples=16.5).fit(X)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_

# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

print('Estimated number of clusters: %d' % n_clusters_)
print('Estimated number of noise points: %d' % n_noise_)
print("Homogeneity: %0.3f" % metrics.homogeneity_score(labels_true, labels))
print("Completeness: %0.3f" % metrics.completeness_score(labels_true, labels))
print("V-measure: %0.3f" % metrics.v_measure_score(labels_true, labels))
print("Adjusted Rand Index: %0.3f"
      % metrics.adjusted_rand_score(labels_true, labels))
print("Adjusted Mutual Information: %0.3f"
      % metrics.adjusted_mutual_info_score(labels_true, labels))
print("Silhouette Coefficient: %0.3f"
      % metrics.silhouette_score(X, labels))

# #############################################################################
# Plot result
import matplotlib.pyplot as plt

# Black removed and is used for noise instead.
unique_labels = set(labels)
colors = [plt.cm.Spectral(each)
          for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = (labels == k)

    xy = X[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=10)

    xy = X[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=6)

plt.title('Estimated number of clusters: %d' % n_clusters_)
plt.show()
```

Note that DBSCAN is used

```python
y_pred = db.labels_.astype(np.int)
y_pred
```

Now the data has been successfully seperated into 2 clusters.

***Supervised learning: Logistic regression***

Using the manually assigned labels in `group` as the true label, the following is to test the effectiveness of Logistic regression. Transform features into $\{x, y, x^2, y^2\}$, then randomly seperate the sample into training set and testing set. 

```python
from sklearn import linear_model
from sklearn.metrics import (precision_score, recall_score, accuracy_score, f1_score)
from sklearn.model_selection import train_test_split


x2 = scatter_x**2
y2 = scatter_y**2
combined = np.vstack((scatter_x, scatter_y, x2, y2)).T
X_train, X_test, y_train, y_test = train_test_split(
    combined, group, test_size=0.95, random_state=42)

logreg = linear_model.LogisticRegression()
logreg.fit(X_train,y_train)
y_pred = logreg.predict(X_test)
print('Accuracy score :', accuracy_score(y_test, y_pred))
print('Precision score :', precision_score(y_test, y_pred))
print('Recall score :', recall_score(y_test, y_pred))
print('F1 score :', recall_score(y_test, y_pred))
```

Random seed is set to 42 so it will be easier to reproduce results. Note using transformed feature $\{x, y, x^2, y^2\}$ the F1 score is very high even using only 5 percent of the dataset as training set. When test_size is any number less than 90 percent of the dataset, the F1 score becomes 1.

```python
colors = np.array(['#ff7f00', '#4daf4a'])
sca0 = plt.scatter(X_train[:, 0], X_train[:, 1], color = '#984ea3', label = 'Training set')
sca1 = plt.scatter(X_test[:, 0], X_test[:, 1], color = colors[y_pred], label = 'Testing set')
plt.scatter([],[],color = '#ff7f00', label = 'Testing set')
#plt.legend()
plt.legend(loc='upper right', fontsize=12)
plt.show()
```

When using larger percent of test sample size, the result is exactly the same as the true result.
